```python
import subprocess
import pandas as pd
import ast
from rdkit import Chem
from rdkit.Chem import Draw
from IPython.display import display, HTML

smiles_list = [
    "C1=CC2=C(C=C1O)C(=CN2)CCN", 
    "CC1=C(C(=NN1C)C)C2=CC=CC(=C2)CCN(C)C", 
    "C1=CC2=C(C=C1C(=O)N)C(=CN2)CCN", 
    "CC1=C(C(=NN1C)C)C2=CC=CC3=C2CC[C@@H](C3)N(C)C", 
    "COC1=CC2=C(C=C1)NC=C2CCN", 
    "CSC1=CC=CC=C1N2CCN(CC2)CCCCCC(=O)NC3CCCC4=CC=CC=C34", 
    "C1CN(CCN1CCCCCN2C3=CC=CC=C3NC2=O)C4=CC=C(C=C4)F",
    "CC1CCN(CC1)CC[C@H]2CCCN2S(=O)(=O)C3=CC=CC(=C3)O"
]

results = []

#loop over SMILES + run external script w subprocess
for smiles in smiles_list:
    try:
        result = subprocess.run(
            ["/opt/anaconda3/envs/biasnet/bin/python", "run_biasnet.py", "--smiles", smiles],
            capture_output=True,
            text=True
        )
        
        #raw output
        raw_output = result.stdout.strip()

        #stdout contains final results ??
        if "Final results:" in raw_output:
            results_part = raw_output.split("Final results: ")[1].split("\n")[0].strip()
            parsed_output = ast.literal_eval(results_part)
            smiles_result = parsed_output.get(smiles, {})
            results.append({
                "SMILES": smiles,
                "Bias prediction": smiles_result.get('Prediction'),
                "Confidence": smiles_result.get('Confidence'),
                "GPCR binding prediction": smiles_result.get('GPCR_Prediction')
            })
        else: #for unexpected output 
            results.append({
                "SMILES": smiles,
                "Bias prediction": None,
                "Confidence": None,
                "GPCR binding prediction": None
            })
    except Exception as e:
        print(f"Error processing SMILES {smiles}: {e}")
        results.append({
            "SMILES": smiles,
            "Bias prediction": None,
            "Confidence": None,
            "GPCR binding prediction": None
        })

#convert stored result into df
df = pd.DataFrame(results)

#generate 2D structure + add to df
images = []
for smiles in df["SMILES"]:
    mol = Chem.MolFromSmiles(smiles)  #convert SMILES to RDKit molecule
    if mol:
        img = Draw.MolToImage(mol, size=(300, 300))  #draw 2D image
        images.append(img)
    else:
        images.append(None)  #for invalid SMILES

df["2D Structure"] = images

#2d structure first col
columns = ["2D Structure"] + [col for col in df.columns if col != "2D Structure"]
df = df[columns]

#output the 2D structure as an image using html rather than <image>
def render_images_with_df(df):
    html = """
    <table border="1" style="border-collapse: collapse;">
        <thead>
            <tr>
                <th>2D Structure</th>
                <th>SMILES</th>
                <th>Bias Prediction</th>
                <th>Confidence</th>
                <th>GPCR Binding Prediction</th>
            </tr>
        </thead>
        <tbody>
    """
    for _, row in df.iterrows():
        smiles = row['SMILES']
        bias_prediction = row['Bias prediction']
        confidence = row['Confidence']
        gpcr_binding = row['GPCR binding prediction']
        
        img_tag = ""
        if row['2D Structure']:
            from io import BytesIO
            import base64
            buffer = BytesIO()
            row['2D Structure'].save(buffer, format="PNG")
            img_data = base64.b64encode(buffer.getvalue()).decode("utf-8")
            img_tag = f'<img src="data:image/png;base64,{img_data}" width="150"/>'
        
        html += f"""
        <tr>
            <td>{img_tag}</td>
            <td>{smiles}</td>
            <td>{bias_prediction}</td>
            <td>{confidence}</td>
            <td>{gpcr_binding}</td>
        </tr>
        """
    html += "</tbody></table>"
    display(HTML(html))

render_images_with_df(df)
```
